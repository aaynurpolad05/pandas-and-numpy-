```python
#EX8-Data Cleaning
df_missing = pd.DataFrame({
    'A':[1,2,np.nan,4],
    'B':[5,np.nan,np.nan,8],
    'C':[10,11,12,13]
})
df_missing.isnull()
```

```python
df_missing.dropna(axis = 'rows', how= 'any')
```

```python
df_missing.reset_index(drop=True).fillna(0, axis= 'rows')
```

```python

df_missing.fillna(df_missing.mean())
```
